```python
import numpy as np
import pandas as pd
import random
from scipy.stats import pearsonr
dir_data = 'C:/study/Dropbox/research/depression/code/data/' # home win10
# dir_data ='C:/Users/yliu456/Dropbox/research/depression/code/data/' #lab
# dir_data = '/home/roy/Dropbox/research/depression/code/data/' # home linux
file = 'studentlife_datafull_v5_withLable.csv'
dataset =  pd.read_csv(dir_data+file)
def minmax_normalize(df): # this min_max norm can apply dataset with NULL value
    columns=df.columns[1:-1]  # assume index and label locates in columns position 1 and -1
    for col in columns:
        df[col]= (df[col]-df[col].min())/(df[col].max()-df[col].min())
#     df.dropna(how='any', inplace=True, axis=1) # drop columns with null value
    return df
data= minmax_normalize(dataset)
data= dataset.fillna(0)
```

```python
#  use covariance, corelationship criteria for feature selection
y= data.iloc[:,-1]

pearson = {}
for i in range(1,88):
    x= data.iloc[:,i]
#     print(data.columns[i],':',np.cov(x,y))
    co, p = pearsonr(x,y)    
    pearson[data.columns[i]]=[co,p]
#     if p <= 0.05:
#         pearson[data.columns[i]]=[co,p]
# X= np.stack((x,y),axis=0)
# cov1 = np.cov(X)
# cov = np.cov(x,y)
pearson
```

```python
cols= ["feature","r","P-value"]
pear_list=pd.DataFrame(columns= cols)
for key, value in pearson.items():
#     print(key,',',value[0],',',value[1])
    pear_list=pear_list.append(pd.Series([key,value[0],value[1]], index= cols), ignore_index= True)
pear_sort = pear_list.sort_values(by=['P-value'])
pear_sort.to_csv(dir_data+'pearson_result.csv')
pear_list.to_csv(dir_data+'pearson_list.csv')
pear_list
```
